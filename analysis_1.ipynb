```python
report['Total']=report[['ENGLISH','HINDI','SCIENCE','MATHS','EVS']].sum(axis=1) #adding a new column
```

```python
report
```

```python
report['Total']/5
```

```python
report['Percentage']=report['Total']/5
```

```python
report
```

```python
report[['NAME','Percentage']].plot(kind='bar',x='NAME',y='Percentage',figsize=(15,5),title='Percentage of Marks',color=['yellow','blue'])
```

```python
report.ENGLISH.plot(kind='hist',
figsize=(10,6),
title='MARKS DISTRIBUTION IN ENGLISH',
bins=15,
color='green',
alpha=.5)
```

```python
report[['ENGLISH','HINDI']].plot(
    kind='hist',
    figsize=(10,6),
    title='MARKS DISTRIBUTION IN ENGLISH',
    bins=15,
    color=['green','red'],
    alpha=.5
)
```

```python
report.plot(
    kind='line',
    x='NAME',
    y=['ENGLISH','HINDI','SCIENCE','MATHS','EVS'],
    rot=45,  #student name rotation
    figsize=(25,5),
    xticks=range(15), #gives all the names(small lines)
    marker='o',
    linewidth=3,
    linestyle='-.'
    )
```

```python
report.plot(
    kind='area',
    x='NAME',
    y=['ENGLISH','HINDI','SCIENCE','MATHS','EVS'],
    rot=45,  #student name rotation
    figsize=(25,5),
    xticks=range(15), #gives 
    stacked=False
    )
```
